```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

# Dataset

```python
iris = load_iris()
X,y = iris.data, iris.target
```

```python
y
```

```python
np.shape(X)
```

```python
np.shape(y)
```

### Creating the decision tree

```python
clf = DecisionTreeClassifier()
```

```python
clf = clf.fit(X, y)
```

```python
clf.predict([[6.1, 2.7, 3.9, 1.2]])
```

### Visualizing the decision tree

```python
plot_tree(clf)
```

```python
plt.figure(figsize = (15,12))
plot_tree(clf, filled = True,
         feature_names = ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width"],
         class_names = ["Sentosa", "Versicolor", "Virginica"])
plt.show()
#Left side is True and Right side is False
# if x[3] = Petal width < = 0.8 then it is almost certainly Iris Sentosa
```

```python
DecisionTreeClassifier(criterion = "entropy")
```

### Pruning a Tree

#### Minimal Cost Complexity Pruning
The algorithm measures the smalled contribution measured in the cost complexity metric and cuts it off from the tree <br>
Depends on a hyper parameter called Alpha: Defines the pruning strength <br>
In scikit-learn it is applied as DecisionTreeClassifier(ccp_alpha = 0)
0 = default value - no pruning <br>
Greater than 1 = pruning will be very severe <br>
Reasonable values are 0.1, 0.01 or 0.001

```python
clf1 = DecisionTreeClassifier(ccp_alpha = 0.1)
```

```python
clf1 = clf1.fit(X, y)
```

```python
clf1.predict([[6.1, 2.7, 3.9, 1.2]])
```

```python
plt.figure(figsize = (15,12))
plot_tree(clf1, filled = True,
         feature_names = ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width"],
         class_names = ["Sentosa", "Versicolor", "Virginica"])
plt.show()
#Left side is True and Right side is False
# if x[3] = Petal width < = 0.8 then it is almost certainly Iris Sentosa
```

## Random Forest
It gives very good accuracy for a non-neural network model for both regression and classification. <br>
It is a type of ensemble model which uses multiple learning algorithms to obtain better predictive performance than any of the constituent learning algorithms alone.<br>
## Bootstrapping
Uniformly sampling from the original dataset with replacement. Each data point we move is replaced in the original data set. and hence there is a chance that the same data point can be selected again.<br>
For large datasets it is expected that 1 - (1/e) times the total number of data points are unique. i.e. about 63% of the sample should be unique and the rest 37% can be duplicate. This is called Bootstrap sample.<br> 
## Bootstrap Aggregating/ Bagging
Bagging is the means of taking the highest number of occurences of a result as the final result in a random forest model. <br>Random forest does not only create a model based on random samples, it also does not consider all the features at the same time. The number of features considered is a hyper parameter but usually it is the square root of the total number of features.

Test the performance of the model

```python
test_pred = clf.predict(test_input)
```

```python
print(classification_report(test_targets, test_pred))
```
