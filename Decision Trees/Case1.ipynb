```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

#from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
raw_data = pd.read_csv(r"C:/Users/justi/OneDrive/STUDY/Python/Satelite_dataset/sat_train.csv")
```

```python
raw_data.head()
```

```python
inputs = raw_data.iloc[:,:-1]
targets = raw_data.iloc[:,-1]
```

```python
train_input, test_input, train_targets, test_targets = train_test_split(inputs, targets, test_size= 0.2, random_state = 365)
```

```python
train_input.describe()
```

```python
train_targets.describe()
```

```python
clf = DecisionTreeClassifier(random_state = 365, ccp_alpha = 0.02)
```

```python
clf.fit(train_input, train_targets)
```

```python
clf.predict([train_input.iloc[125]])
```

```python
test_pred = clf.predict(test_input)
```

```python
print(classification_report(test_targets, test_pred))
```

```python
from sklearn import tree
plt.figure(figsize = (10,10))
tree.plot_tree(clf, filled= True)
plt.show()
```
